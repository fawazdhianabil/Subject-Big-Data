###**FIX THE OUTLIERS**

This technique uses the IQR scores calculated earlier to remove outliers. The rule of thumb is that anything not in the range of (Q1 - 1.5 IQR) and (Q3 + 1.5 IQR) is an outlier, and can be removed. 

````python
#Count Skor IQR
sorted(df_new)
Q1=df_new.quantile(0.25)
Q3=df_new.quantile(0.75)
IQR=Q3-Q1
print(IQR)
````

The first line of code below removes outliers based on the IQR range and stores the result in the data frame 'df_cl'. <br>

````python
df_cl = df_new[~((df_new < (Q1 - 1.5 * IQR)) |(df_new > (Q3 + 1.5 * IQR))).any(axis=1)]
df_cl
````

The second line prints the shape of this data, which comes out to be 3079 observations of 7 variables. This shows that for our data, a lot of records get deleted if we use the IQR method.

````python
print(df_cl.shape)
````

````python
df_2=df_cl.select_dtypes(include=['float64', 'int64'])
# To check the outliers we use boxplots.
plt.figure(figsize=(10,15))
for i, col in enumerate(list(df_2.columns.values)):
    plt.subplot(4,3,i+1)
    df_cl.boxplot(col)
    plt.grid()
    plt.tight_layout()
````

There are no any outliers aggain.

Skewness is a measure of the asymmetry in the distribution of values. Skewness can be positive, negative, and zero. <br>

A positive Skewness means that the tail of the distribution is to the right of the highest value. This means that most of the distributions are in low values. A negative Skewness means that the tail of the distribution is on the left, indicating that most of the values â€‹â€‹are on the right side of the curve. While skewness is zero, it means that the value is symmetrically distributed, with the distance between the tails of the distribution on the right and the left being the same.

image.png

Kurtosis is an indicator to show the degree of tailedness. The greater the kurtosis value, the sharper the curve. <br>

Skewness and kurtosis can indicate the condition of the distribution or distribution of data. The ideal condition is when the data is normally distributed, when skewness is 0 and kurtosis is 3. The farther from the ideal condition, the more non-ideal or unevenly distributed the data is.
