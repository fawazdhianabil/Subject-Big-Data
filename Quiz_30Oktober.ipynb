```python
#8. Menampilkan 10 Provinsi terbanyak dalam dataframe
dff= df['Provinsi'].value_counts()
dff.head(10)
pd.DataFrame(dff)
```

```python
#9. Menampilkan data dengan alamat mengandung kata jl.
df_alamat=df[['Nama Lengkap Siswa','Alamat','Nomor Induk Siswa Nasional (NISN)']]
df_alamat[df_alamat.Alamat.str.contains('Jl.')]
```

```python
#10. Menampilkan data dengan per provinsi dengan menampilkan rata-rata nilai semester 4, 5, dan 6 tampil terurut dari paling tinggi ke paling rendah
dfff=df[['Provinsi','Nilai Rata-rata Semester 4 (Nilai Pengetahuan)','Nilai Rata-rata Semester 5 (Nilai Pengetahuan)','Nilai Rata-rata Semester 6 (Nilai Pengetahuan)']]
df_m=dfff.groupby(['Provinsi'],sort=True).mean()
df_m.sort_values(['Nilai Rata-rata Semester 4 (Nilai Pengetahuan)'],ascending=False)
```
