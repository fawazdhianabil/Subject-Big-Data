## Scater Plot

Scatter Plot adalah plot berupa titik-titik data dalam dimensi 2 ataupun 3. Tujuan melakukan plot dalam bentuk ini adalah melihat sebaran data, melihat pola-pola sebaran data dalam bentuk kelompok, bisa juga untuk melihat pola keterkaitan(korelasi), ataupun kedekatan antar titik data.

Contoh kasus untuk pengujian kekuatan hubungan antara 2 variabel antara lain :
*   Hubungan antara kecepatan Mesin dengan Kualitas Produk.
*   Hubungan antara Jumlah Tenaga Kerja dengan Output yang dihasilkan.
*   Hubungan antara Jumlah Jam kerusakan mesin dengan tingkat kecacatan yang terjadi.
*   Hubungan antara Total Jam Lembur dengan tingkat absensi Tenaga Kerja.
*   Hubungan antara Absensi dengan tingkat kerusakan produk.


image.png

Untuk membentuk scatter plot, kita bisa menggunakan fungsi ```scatter()```.

Fungsi scatter() membuat plot dengan tanda titik untuk mempresentasikan datanya. dibutuhkan 2 array dengan panjang yang sama untuk membentuk scatter plot



````
#Membuat Plot Scatter dengan Matpolib
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
````

````
#Membuat Data
np.random.seed(10)
x = np.random.rand(100)
y = np.random.randint(10, 70, 100)
cat = ['Bahasa', 'IPA', 'IPS']
jurusan = np.random.choice(cat, 100)

df = pd.DataFrame(dict(X=x, Y=y, Jurusan=jurusan))
df.head(10)
````

Atribut dalam Scatterplot

Syntax: Axes.scatter(self, x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=, edgecolors=None, *, plotnonfinite=False, data=None, **kwargs)

````
#Membuat Scatterplot dengan Matplotlib
colors = {'Bahasa':'red', 'IPA':'blue', 'IPS':'yellow'}

fig, axes = plt.subplots()
axes.scatter(df['X'], df['Y'], c=df['Jurusan'].map(colors))
axes.set_title('Banyak Siswa dan Tingkat Stress')
axes.set_xlabel('Tingkat Stress')
axes.set_ylabel('Banyak Siswa')
plt.show()
````

````
#Membedakan Titik Berdasarkan Warna
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.show()
````

````
#Membedakan Titik Berdasarkan Ukuran
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes= np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)
plt.show()
````

````
#Membedakan Titik Berdasarkan Alpha
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
````

````
#Menggabungkan Warna, Ukuran dan Alpha
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.show()
````

````
#Cara Meresize Scatterplot
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))
plt.gcf().set_size_inches((10, 10))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.show()
````

## Bar Plot

Bar plot atau diagram batang, diagra ini cocok untuk membandingkan nilai data yang cukup banyak. Berfungsi untuk melihat kenaikan atau penurunan data.

### Vertical Bar Plot

````
#Membuat dataset
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

#Membuat Plot Bar vertikal
plt.bar(x,y, color ='green', width = 0.5)
plt.xlabel("Huruf")
plt.ylabel("Jumlah")
plt.title("Contoh Bar Plot")
plt.show()
````

### Horizontal Bar Plot

Jika contoh sebelumnya adalah membuat bar plot vertikal , sekarang kita akan membuat bar plot horizontal.

Pembuatan bar plot horizontal tidak jauh berbeda dengan bar plot vertikal. Yang membedakan tentunya hanya method yang digunakan saja. Untuk membuat bar plot horizontal kita menggunakan method *barh()*. Perhatikan kode berikut.

````
#Membuat Plot Bar horizontal
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
````

## Pie Plot

Diagram lingkaran mengacu pada grafik melingkar yang dipecah menjadi beberapa segmen yaitu irisan. Pie plot digunakan untuk menyajikan data dalam bentuk persentase, yang mana setiap potongan pie berisikan data dengan ukuran tertentu. Visualisasi diagram lingkaran fokus untuk menampilkan perbandingan data. Lebih baik digunakan untuk jumlah variabel yang sedikit. Secara visual pie plot menarik untuk penyajian data serta memudahkan pembaca dalam memahami data tersebut.

### Diagram pie sederhana

````
values = [2500, 1500, 1053, 1200]

plt.pie(values)
plt.show()
````

Seperti yang kita lihat, diagram pie menggambar satu bagian (disebut irisan) untuk setiap nilai dalam array (dalam hal ini [2500, 1500, 1053, 1200]).

Secara default, plotting irisan pertama dimulai dari sumbu x dan bergerak berlawanan arah jarum jam:

![](https://drive.google.com/uc?export=view&id=1DL2LHo9KgcFsYVGT7A76D-GsesKu6lO5)

### Label

Tambahkan label ke diagram pie dengan menggunakan parameter `labels`.

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']

plt.pie(values, labels=label)
plt.show()
````

### Sudut Awal

Seperti yang disebutkan, sudut awal default berada pada sumbu x, tetapi kita dapat mengubah sudut awal dengan menentukan parameter `startangle`.

Parameter `startangle` ditentukan dengan sudut dalam derajat, sudut default adalah 0:

![](https://drive.google.com/uc?export=view&id=15CvRYeYC79lQ4nnuWznkdqwf60NhfaV9)

Contoh: Mulai irisan pertama pada 90 derajat:

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']

plt.pie(values, labels=label, startangle = 90)
plt.show()
````

### Explode

Parameter `explode` memungkinkan untuk membuat salah satu irisan terlihat lebih menonjol.

Parameter `explode`, jika ditentukan (bukan None) harus berupa array dengan satu nilai untuk setiap irisan.

Setiap nilai mewakili seberapa jauh dari pusat setiap irisan ditampilkan.

Contoh: Tarik irisan Matematika 0,1 dari bagian tengah pie

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']
myexplode = [0.1, 0, 0, 0]

plt.pie(values, labels=label, explode = myexplode)
plt.show()
````

### Bayangan

Tambahkan bayangan pada diagram pie dengan mengatur parameter `shadow` menjadi **True**.

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']
myexplode = [0.1, 0, 0, 0]

plt.pie(values, labels=label, explode = myexplode, shadow = True)
plt.show()
````

### Warna

Kita dapat mengatur warna setiap irisan dengan parameter `colors`.

Parameter `colors`, jika ditentukan, harus berupa array dengan satu nilai untuk setiap irisan.

Contoh:
Tentukan warna baru untuk setiap irisan

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']
mycolors = ["black", "hotpink", "b", "g"]

plt.pie(values, labels=label, colors = mycolors)
plt.show()
````

Kita dapat menggunakan nilai warna Heksadesimal, salah satu dari 140 nama warna yang didukung, atau salah satu pintasan berikut untuk menentukan warna diagram pie.

*   r  Merah
*   g  Hijau
*   b  Biru
*   c  Cyan
*   m  Magenta
*   y  Kuning
*   k  Hitam
*   w  Putih

### Legend

Untuk menambahkan penjelasan pada setiap irisan, gunakan fungsi `legend()`

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']

plt.pie(values, labels=label)
plt.legend()
plt.show()
````

### Legend Dengan Header

Untuk menambahkan header pada legend, tambahkan parameter **title** pada fungsi legend.

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']

plt.pie(values, labels=label)
plt.legend(title = "Program Studi:")
plt.show()
````

### Nilai Persen

Untuk menambahkan nilai persen pada setiap irisan, gunakan `autopct`

````
values = [2500, 1500, 1053, 1200]
label = ['Matematika','Biologi','Fisika','Sistem Informasi']
myexplode = [0.1, 0, 0, 0]

plt.pie(values, labels=label, explode = myexplode, shadow = True, autopct='%2.1f%%')
plt.show()
````

### Implementasi Pie Plot Menggunakan Super Store Dataset

Implementasi pie plot menggunakan data pada kolom category dan sales untuk melihat sales pada masing-masing category.

````
# Membuat list category
list_category=sorted(df_store['category'].unique())

# Mengelompokkan category dengan sales
df_1=df_store.groupby(df_store['category']).agg({'sales':sum})
df_1.head()

# Membuat dataframe category untuk melihat masing-masing salesnya
df_2= pd.DataFrame(
    { 'category'  : list_category,
      'sales'     : df_1['sales']
    
    })
df_2
````

````
# Membuat pie plot
plt.pie(df_2['sales'], labels=df_2['category'], explode = (0,0,0.1), shadow = True, autopct='%1.1f%%')
plt.show()
````

## Violin Plot

Plot Violin adalah representasi statistik dari data numerik

````
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 20)
collectn_2 = np.random.normal(80, 30, 20)
collectn_3 = np.random.normal(90, 20, 20)
collectn_4 = np.random.normal(70, 25, 20)

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

pd.DataFrame(data_to_plot)
data_to_plot
````

````
# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data_to_plot)
plt.show()
````
